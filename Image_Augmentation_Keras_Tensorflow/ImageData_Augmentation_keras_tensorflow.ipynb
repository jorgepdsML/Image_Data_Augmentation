#### IMAGE DATA AUGMENTATION USING KERAS & TENSORFLOW

Tecnica que permite poder expandir el dataset de entrenamiento disponible mediante el uso de transformaciones sobre las imagenes con el fin de mejorar el rendimiento de los modelos de aprendizaje automatico .El uso del aumento de datos mediante keras y tensorflow permite poder aprovechar de mejor forma los recurso de memoria debido a que se estaran creando nuevas imagenes en etapa de entrenamiento sin tener que requerir mas espacio de memoria para guardar las imagenes modificadas.

```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
import numpy as np
```

```python
#cargar imagen
img=load_img("izumi.jpg")

#convertir a array
img=img_to_array(img)

plt.imshow(np.uint8(img))

#agregar una dimension en axis=0
x=np.expand_dims(img,axis=0)

print(x.shape)
```

### ImageDataGenerator

Esta clase permite poder generar batches de tensores de imagenes la cual posee argumentos que determinan las transformaciones que se realizaran a las imagenes de entrada.

**horizontal_flip**: Argumento del tipo bool que cuando es True , se realizara una transformacin sobre la imagen que cambiara las posiciones de los pixeles de la imagen de entrada en sentido horizontal.

**rotation_range**:Argumento del tipo entero que determina el rango de rotacin que se aplicara a la imagen de entrada.

**brightness_range**:Argumento del tipo tupla (min,max) que determina el valor que se escogera para la modificacin del brillo sobre la imagen de entrada

**zoom_range**: Argumento flotante o tupla que determina el rango de zoom que se aplicara a la imagen de entrada.

#### iterador
un iterador es un objeto de python que contiene una cantidad de elementos disponibles y que puede ser iterado , es decir se puede realizar un recorrido sobre todos sus elementos.

todo iterador posee los siguientes mtodos __iter__() y __next__().

la funcin **next** permite poder acceder al siguiente elemento de un iterador , esto lo hace invocando internamente el mtodo __next__() del objeto iterador. 

cuando ya no existe mas elementos entonces se originara una excepcin **StopIteration**.

#### iterable 
un objeto iterable se caracteriza por tener definido el mtodo **iter** que retorna un iterador.

un interable puede ser una lista, range, tupla , diccionarios . entre otras.

**braintels=["IA","Image_Processing","Embedded_systems","BRAINTELS"]**

```python
#crear objeto iterable (lista) 
braintels=["IA","Image_Processing","Embedded_systems","BRAINTELS"]
#conseguir el iterador del objeto braintels mediante la funcin iter()
iterador=iter(braintels)
#realizar un recorrido sobre los elementos del iterador mediante la funcin next()
print(next(iterador)) # "IA" 
print(next(iterador)) # "Image_Processing"
print(next(iterador)) #"Embedded_systems"
print(next(iterador)) #"BRAINTELS"
#si se coloca de nuevo la funcin next(iterador) entonces originara una exception StopIteration
```

### Flipping images

```python
#-----------------crear objeto de la clase ImageDataGenerator

#-----------------transformacin horizontal_flip---------
dato_generador=ImageDataGenerator(horizontal_flip=True)

#----------------Crear iterator mediante el mtodo flow
#iterador
it = dato_generador.flow(x, batch_size=1)
```

```python
plt.figure(figsize=(7,5))
plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.20)
for i in range(9):
	# definir subplot
	plt.subplot(3,4,i+1)
	# acceder a un elemento del iterador it , devuelve la imagen modificada de forma aleatoria
	batch= next(it)
	# convertir e entero sin signo de 8 bits
	image = batch[0].astype('uint8')
	# mostrar la imagen
	plt.imshow(image)
	plt.axis("off")
# show the figure
pyplot.show()
```
